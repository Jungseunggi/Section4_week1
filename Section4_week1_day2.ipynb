* **순전파(forward propagation)**는 위 그림과 같이 입력층부터 출력층까지 가는 과정을 말함
* 순전파은 뉴럴 네트워크 모델의 입력층부터 출력층까지 순서대로 변수들을 계산하고 저장하는 것을 의미
* **역전파(back propagation)**는 순전파의 역순
* 역전파는 뉴럴 네트워크의 파라미터들에 대한 경사(gradient)를 계산하는 방법
* 역전파는 쉽게 말해서 가장 가까운 레이어부터 가중치를 수정하면서 오차를 줄임
* 역전파의 학습은 미분의 **'chain rule'**을 따름
* 따라서 딥러닝 모델을 학습시킬 때, 순전파와 역전파는 상호 의존적
* 순전파, 역전파를 통해 갱신하는 과정을 **Iteration(이터레이션**)이라고 함
* 예를 들면 우리가 음식을 맛보는데 간이 싱거우면 소금을 더 넣고 짜면 물을 더 넣으면서 간을 맞추는 과정이라 볼 수 있음

##**1.2 경사하강법(Gradient Descent)**

image.png

이미지 출처 : https://winder.ai/402-optimisation-and-gradient-descent/

* Gradient Descent 방법은 1차 미분계수를 이용해 함수의 최소값을 찾아가는 iterative한 방법
* 실제 미분계수를 계산하는 과정을 컴퓨터로 구현하는 것에 비해 gradient descent는 컴퓨터로 비교적 쉽게 구현
* 데이터 양이 매우 큰 경우 gradient descent와 같은 iterative한 방법을 통해 해를 구하면 계산량 측면에서 더 효율적

###**1.2.1 Step size**

image.png

이미지 출처 : https://angeloyeo.github.io/2020/08/16/gradient_descent.html

* step size가 큰 경우 한 번 이동하는 거리가 커지므로 빠르게 수렴할 수 있다는 장점
* step size를 너무 크게 설정해버리면 최소값을 계산하도록 수렴하지 못하고 함수 값이 계속 커지는 방향으로 최적화가 진행될 수 있음
* 또, 한편 step size가 너무 작은 경우 발산하지는 않을 수 있지만 최적의 [Math Processing Error]를 구하는데 소요되는 시간이 오래 걸린다는 단점

* 옵티마이저는 가중치 및 학습률과 같은 신경망의 속성을 수정하는 함수 또는 알고리즘
* 따라서 전체 손실을 줄이고 정확도를 높이는 데 도움
* 쉽게 말하면 경사를 내려가는 방법

###**1.3.1 배치 경사 하강법 (Batch Gradient Descent, BGD)**

image.png

이미지 출처 : https://skyil.tistory.com/68

* 여기서 배치란 GPU가 한번에 처리하는 데이터의 묶음을 의미
* 전체 학습 데이터를 하나의 배치로(배치 크기가 n) 묶어 학습
* 전체 데이터를 통해 학습시키기 때문에, 업데이트 횟수가 적음(1 Epoch 당 1회 업데이트)
* 전체 데이터를 모두 한 번에 처리하기 때문에, 메모리가 가장 많이 필요
* 항상 같은 데이터 (전체 데이터)에 대해 경사를 구하기 때문에, 수렴이 안정적

###**1.3.2 확률적 경사 하강법(Stochastic Gradient Descent, SGD)**

image.png

이미지 출처 : https://skyil.tistory.com/68

* 확률적 경사 하강법에서는 각 반복에 대해 전체 데이터 세트를 가져오는 대신 데이터 배치를 무작위로 선택
* 각 반복에 대해 일괄 처리를 사용하기 때문에 알고리즘이 사용하는 경로는 BGD보다 노이즈가 심함(위 이미지 참고)
* 따라서 SGD는 로컬 최소값에 도달하기 위해 더 많은 수의 반복을 사용 -> 반복 횟수의 증가로 인해 전체 계산 시간이 증가
* 결론은 데이터가 방대하고 계산 시간이 필수 요소인 경우 BGD보다 SGD 선호

###**1.3.3 모멘텀(Momentum)**

* 앞에 BGD보다 SGD는 노이즈가 심함 따라서 global minimum에 도달하려면 많은 반복이 필요 따라서 많은 시간 소요
* 이 문제점을 해결하려고 나온게 Momentum

image.png

이미지 출처 : https://www.analyticsvidhya.com/blog/2021/10/a-comprehensive-guide-on-deep-learning-optimizers/

* 위 이미지(왼쪽=SGD, 오른쪽=Momentum)를 보면 운동량을 사용하면 더 짧은 시간에 수렴에 도달하는 데 도움이 된다는 것을 확인 가능
* 그러나 모멘텀을 증가시키면 최적의 최소값을 통과할 가능성 증가, 정확도가 떨어지고 많은 노이즈 발생 할 수 있음

###**1.3.4 미니 배치 확률적 경사하강법(Mini-Batch Stochastic Gradient Descent,MSGD)**

image.png

이미지 출처 : https://skyil.tistory.com/68

* 딥러닝 라이브러리 등에서 SGD를 얘기하면 최근에는 대부분 이 방법을 의미
* 전체 데이터를 batch_size개씩 나눠 배치로 학습(배치 크기를 사용자가 지정)
* BGD보다 계산량이 적음
* Shooting이 적당히 발생(Local Minima를 어느정도 회피할 수 있음)

###**1.3.5 Adagrad(Adaptive Gradient Descent)**

* 반복마다 다른 학습률을 사용
* 학습률의 변화는 훈련 중 매개변수의 차이에 따라 달라짐
* 학습률을 수동으로 수정할 필요가 없음
* 학습률을 공격적이고 단조롭게 감소, 따라서 학습률이 극도로 작아지는 시점이 있을 수 있음

###**1.3.6 RMS Prop(Root Mean Square)**

* 각 가중치에 대해 단계 크기를 개별적으로 조정하는 기울기의 부호를 사용
두 개의 기울기는 먼저 부호에 대해 비교
* 같은 부호 -> 올바른 방향으로 가고 있는 것이고 따라서 small fraction(이건 어떤 의미로 쓰인건지 모르겠네요ㅜ)만큼 step size를 증가
* 반대 부호 -> step size를 줄이고 다음 step size를 제한
* 이해한바로는 기울기가 같은 부호이면 step size를 증가 시키고 다르면 감소시킨다는 의미인거 같음

###**1.3.7 Adam**

* 최적화 알고리즘은 훈련 중 네트워크 가중치를 업데이트하기 위해 확률적 경사하강법을 추가로 확장한 것
* Adam은 Adagrad 및 RMS prop 알고리즘의 기능을 가지고 있음
* SGD에서 훈련을 통해 단일 학습률을 유지하는 것과 달리 Adam은 각 네트워크 가중치에 대한 학습률을 개별적으로 업데이트
* 구현하기 쉽고 실행 시간이 더 빠르며 메모리 요구 사항이 낮고 다른 최적화 알고리즘보다 조정이 덜 필요

###**1.3.8 배치 사이즈**


* 보통 2의 n승으로 지정
* 메모리 크기가 허락한다면 큰 배치 사이즈를 쓰는 것이 학습을 안정적으로 진행
* 가능하면 학습데이터 갯수에 나누어 떨어지도록 지정하는 것이 좋음 
* 마지막 남은 배치가 다른 사이즈라면 해당 배치의 데이터가 학습에 더 큰 비중을 갖게 됨
* 예를 들어, 530 개의 데이터를 100개의 배치로 나누면, 각 배치 속 데이터는 1/100 만큼의 영향력을 가지나 마지막 배치(30개)의 데이터는 1/30의 영향력을 갖게 되어 과평가되는 경향
* 따라서 보통 마지막 배치의 사이즈가 다를 경우 이는 버리는 방법을 사용

##**1.4 epoch, batch-size, iterations**

### **1.4.1 에포크(Epoch)**

* 1 epoch는 전체 학습데이터셋이 한 신경망에 적용되어 순전파, 역전파를 통해 신경망을 한번 통과했다는 의미
* epoch를 높일수록 적합한 파라미터를 찾을 확률이 증가(손실 값 하락)
* 지나치게 높으면 오버피팅

이미지 출처 : https://towardsdatascience.com/epoch-vs-iterations-vs-batch-size-4dfb9c7ce9c9

**하나 이상의 epoch를 사용하는 이유??**

*  우리는 제한된 데이터 세트를 사용하고 있으며 학습과 그래프를 최적화하기 위해 반복 프로세스 인 Gradient Descent 를 사용하고 있음
* 따라서 단일 패스 또는 한 epoch로 가중치를 업데이트하는 것만으로는 충분하지 않음
* 한 에포크(epoch)는 그래프에서 곡선의 과소적합이 될 수 있음

###**1.4.2 배치 사이즈(Batch size)**


* 연산 한 번에 들어가는 데이터 크기

**배치 사이즈 큰거 vs 작은거**

* 배치 사이즈가 너무 큰 경우 한번에 처리해야 할 데이터의 양이 많아짐 따라서 학습 속도 저하, 메모리 부족

* 배치 사이즈가 너무 작은 경우 적은 데이터를 대상으로 가중치를 업데이트함으로 훈련이 불안정
